```python
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
import simpleaudio as sa
import librosa
```

```python
# Importing the audio and printing the details
audio, sr = librosa.load("/content/LJ050-0240.wav",sr=None)
print(f"Sample Rate: {sr} Hz")
print(f"Number of Samples: {len(audio)}")
print(f"Duration: {len(audio)/sr} seconds")
samples = len(audio)
dur = len(audio)/sr
```

```python
#Plotting the wave form
time = np.linspace(0,dur,samples)
plt.figure(figsize=(10, 4))
plt.plot(time,audio, color='blue')
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Waveform of Speech Signal")
plt.show()
```

```python
#Slicing the audiofile to 2 secs
sample_twosec = int(2*sr)
sliced_audio = audio[:sample_twosec]
sliced_time = np.linspace(0,2,len(sliced_audio))

plt.figure(figsize=(10, 4))
plt.plot(sliced_time,sliced_audio, color='black')
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("2 sec sliced Speech Signal")
plt.show()
```

```python
# Normalize the audio
normalized_audio = sliced_audio / np.max(np.abs(sliced_audio))
time_normalized = np.linspace(0, 2, len(normalized_audio))

plt.figure(figsize=(10, 4))
plt.plot(time_normalized,normalized_audio, color='black')
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Normalized Speech Signal")
plt.show()
```

```python
# Amplification
amplified_audio = normalized_audio * 2

# Deamplification
deamplified_audio = normalized_audio * 0.5

plt.figure(figsize=(10, 4))
plt.plot(time_normalized,amplified_audio, color='black')
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Amplified Speech Signal")
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(time_normalized,deamplified_audio, color='black')
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Deamplified Speech Signal")
plt.show()
```

```python
# Up-sampling
upsampled_audio = librosa.resample(normalized_audio, orig_sr=sr, target_sr=sr * 2)

# Down-sampling
downsampled_audio = librosa.resample(normalized_audio, orig_sr=sr, target_sr=sr // 2)

upsamp_time = np.linspace(0, 2, len(upsampled_audio))
downsamp_time = np.linspace(0, 2, len(downsampled_audio))

plt.figure(figsize=(10, 4))
plt.plot(upsamp_time,upsampled_audio, color='black')
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Up Sampled Speech Signal")
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(downsamp_time,downsampled_audio, color='black')
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Down Sampled Speech Signal")
plt.show()
```
